```python
# import library
import pandas as pd

from bs4 import BeautifulSoup
import requests
```

```python
# Request to website and download HTML contents
current_page = 0
url = 'https://myanimelist.net/topanime.php?limit={}'.format(current_page)
req = requests.get(url)
content =  req.text
```

```python
# checks status code
req.status_code
```

```python
content
```

```python
# format using BeautifulSoup to make it readable
soup = BeautifulSoup(content)
```

```python
soup
```

```python
table = soup.find('table')
type(table)
```

```python
table.find_all('td')
```

```python
animes = []

for row in table:
    print(row)
    
```

```python
animes = []

for row in table.find_all('td',{'class':'title al va-t word-break'}):
    anime = {}
    
    anime['name'] = row.find('h3').text
    anime['details'] = row.find_all('div',{'class':'information di-ib mt4'})[0].text.strip()
    anime['score'] = 0
    animes.append(anime)
    
for row in table.find_all('td',{'class':'score ac fs14'}):
    anime={}
    anime['score'] = row.find('span').text
    animes.append(anime)
    
pd.DataFrame(animes).dropna()
```
